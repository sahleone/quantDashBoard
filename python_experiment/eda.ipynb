# Use report from Robinhood to get portfolio

```python
import pandas as pd
```

```python
data =  pd.read_csv('c5f0fca5-716f-5430-95ca-712caefd7b45.csv', on_bad_lines='skip', parse_dates=['Activity Date', 'Process Date', 'Settle Date'])

data = data.dropna(subset=['Description'])

# Now you can safely add columns
data['Settle Year'] = data['Settle Date'].dt.year
data['Settle Month'] = data['Settle Date'].dt.month

# Convert the Amount column to float
## Remove the $ sign
data['Amount_asFloat'] = data['Amount'].replace('[\$,]', '', regex=True)

## Convert (number) to negative
data['Amount_asFloat'] = data['Amount_asFloat'].replace('[\(]', '-', regex=True)
data['Amount_asFloat'] = data['Amount_asFloat'].replace('[\)]', '', regex=True)
## Convert to float
data['Amount_asFloat'] = data['Amount_asFloat'].astype(float)

data.head()
```

## Dividends


```python
# Create an explicit copy to avoid the warning
data_div = data[data['Description'].str.contains('Cash Div')].copy()


data_div.groupby(["Settle Year", "Settle Month"])['Amount_asFloat'].sum()
```
